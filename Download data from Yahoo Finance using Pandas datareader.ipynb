```python
# Install and Import yfinance
!pip install yfinance
import yfinance as yf
```

```python
# Download Times series data for META from 01/01/2014 to 16/08/2025
df_meta = yf.download("META", start="2014-01-01", end="2025-08-16")
print(df_meta) #To print downloaded data
```

```python
#Ploting High variable (Second Column in the df_meta dataset)
df_meta['High'].plot(figsize = (12,4), c = "Yellow") #c means colors where you can choose from different colors options
```

```python
# Now let us Download data for Tesla (TSLA)
df_tesla = yf.download("TSLA") #you can choose a the start and the end using yf.download("TSLA", start = "starting date", end = "ending date")
# Show only first few rows
df_tesla.head(100) # Use head for the first rows
```

```python
# In case you want last rows use Tail
df_tesla.tail()
```

```python
# Now let us download Gold data
import yfinance as yf
df_gold = yf.download("GC=F", start="2024-07-01", end="2025-08-16")
# Show first few rows
df_gold.head(5)
```

```python
# Ploting data
df_gold['High'].plot(xlim = ["2025.01.01", "2025.08.01"],ylim = [2600, 3600], figsize=(12,4),  ls = "-.", c = "red")
```

```python
#Get timeline of data dates
df_gold.index
```

```python
#Subset Share_High data (Second Column)
share_High = df_gold.loc['2025-08-01': "2025-08-16"]['High']
print(share_High)
index = df_gold.loc['2025-08-01': "2025-08-16"].index
print(index)
# Note: in this code we extract only data for High with its index (Date)
```

```python
#Start ploting High with its index
import matplotlib.pyplot as plt
%matplotlib inline
figure, axis= plt.subplots()
plt.tight_layout()
#preventing overlaping
figure.autofmt_xdate()
axis.plot(index, share_High, c= "red", ls= ":")
```

```python
#Datetime index
df_gold.info()
```

```python
#Using this code allow us to change index of data from Date to numbers
df = df_gold.reset_index()
df.info()
```

```python
#Check the resuls of changing the Index
print(df)
```

```python
# Done. Let us download Gold data to an excel file. If the code is correct than write "Thank You"
df.to_csv('Gold.csv', index=False)
print('Thank you')
```
