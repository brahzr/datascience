```python
#Working file info
file_path = r"C:\Users\Brahim\Downloads\statistical-performance-indicator.csv"  #Your csv file path

#Importing and loading DATA
import pandas as pd
df = pd.read_csv(file_path)

#Filter criteria
selected_countries = ["Algeria","Malaysia","Saudi Arabia"]
selected_variable = "Statistical performance indicators (SPI): Overall score (scale 0-100)"  # Choose the variable you need
output_file = "filtered_SPI.xlsx"  # Output file name

#Filter Selected Sata
filtered_df = df[
    df["Entity"].isin(selected_countries)
    ]

#Sort by Entity and Year
filtered_df = filtered_df.sort_values(["Entity", "Year"])

#Display the filtered data
print("Filtered Data:")
display(filtered_df)

#SAVE TO EXCEL
filtered_df.to_excel(output_file, index=False)
```
